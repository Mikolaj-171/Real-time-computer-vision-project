```python
import cv2
from ultralytics import YOLO

MODEL_PATH = r'best.pt' 
model = YOLO(MODEL_PATH)

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Błąd: Nie można połączyć się z kamerą.")
    exit()

print("Kamera uruchomiona! Wciśnij klawisz 'q' w oknie z obrazem, aby wyjść.")

while True:
    success, frame = cap.read()
    
    if not success:
        print("Błąd: Nie udało się pobrać klatki z kamery.")
        break

    frame = cv2.flip(frame, 1)
    results = model.predict(source=frame, stream=True, conf=0.4, verbose=False)

    for result in results:
        annotated_frame = result.plot()

    cv2.imshow('Rozpoznawanie Gestow - YOLOv8', annotated_frame)


    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Zamykanie programu...")
        break

cap.release()
cv2.destroyAllWindows()
```
